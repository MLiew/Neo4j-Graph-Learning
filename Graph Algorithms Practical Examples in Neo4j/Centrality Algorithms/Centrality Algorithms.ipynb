Centrality algorithms are used to understand the importances and roles of particular nodes in a graph. They identify the most important nodes and help us understand group dynamics. In this notebook, we'll use degree centrality, closeness centrality, betweeness centrality, and page rank. 

1. Degree centrality: Measures the number of relationships a node has.
2. Closeness centrality: Calculates which nodes have the shortest paths to all other nodes. 
3. Betweeness centrality: Measures the number of shortest paths that pass through a node.
4. Page rank: Estimates a current node's importance from its linked neighbors and their neighbors.

## Create graph

```python
# Import libraries
import pandas as pd
from neo4j import GraphDatabase
```

```python
# connect to neo4j server
user = "neo4j"
password = "xxx"  # change your own password
driver = GraphDatabase.driver("bolt://localhost:7687", auth=(user, password))
```

```python
# load the csv files of social nodes and transport relationships
with driver.session() as session:
    session.run("""
    LOAD CSV WITH HEADERS FROM "file:///data/social-nodes.csv" AS row
    MERGE (:User {id:row.id})
    """)
    
    session.run("""
    LOAD CSV WITH HEADERS FROM "file:///data/social-relationships.csv" AS row
    MATCH (source:User {id: row.src})
    MATCH (destination:User {id: row.dst})
    MERGE (source)-[:FOLLOWS]->(destination)
    """)
```

```python
# create in-memory graph, set the orientation to "REVERSE" to retrieve people with the most followers
create_memory_graph_query = """
CALL gds.graph.create(
    'social',
    ['User'],
    {
      FOLLOWS: {
      orientation: 'REVERSE'
    }}
)
"""
```

```python
with driver.session() as session:
    session.run(create_memory_graph_query)
```

## Degree Centrality
The Degree Centrality algorithm can be used to find popular nodes within a graph. Degree centrality measures the number of incoming or outgoing (or both) relationships from a node, depending on the orientation of a relationship projection. It can be applied to either weighted or unweighted graphs. In the weighted case the algorithm computes the sum of all positive weights of adjacent relationships of a node, for each node in the graph. Non-positive weights are ignored.

The average degree of a network is simply the total number of relationships divided by the total number of nodes; it can be highly skewed by high degree nodes. The degree distribution is the probability that a randomly selected node willhave a certain number of relationships. 

```python
degree_query = """
    CALL gds.degree.stream('social')
    YIELD nodeId, score
    RETURN gds.util.asNode(nodeId).id AS id, score AS followers
    ORDER BY followers DESC, id DESC
"""
```

```python
with driver.session() as session:
    rows = session.run(degree_query)
    degree_df = pd.DataFrame([dict(record) for record in rows])
```

```python
degree_df
```

Doug has the most followers.

## Closeness Centrality
Closeness centrality is a way of detecting nodes that are able to spread information very efficiently through a graph. The closeness centrality of a node measures its average farness (inverse distance) to all other nodes. Nodes with a high closeness score have the shortest distances to all other nodes.

```python
close_query = """
    CALL gds.alpha.closeness.stream({
      nodeProjection: 'User',
      relationshipProjection: 'FOLLOWS'
    })
    YIELD nodeId, centrality
    RETURN gds.util.asNode(nodeId).id AS user, centrality
    ORDER BY centrality DESC
"""
```

```python
with driver.session() as session:
    rows = session.run(close_query)
    close_df = pd.DataFrame([dict(record) for record in rows])
```

```python
close_df
```

The score represents their closeness to the others within the subgraph but not the entire graph. 

### Closeness Centrality Variation: Harmonic Centrality

Rather than summing the distances of a node to all other nodes, the harmonic centrality algorithm sums the inverse of those distances. This enables it deal with infinite values.

The raw harmonic centrality for a node is calculated using the following formula:

raw harmonic centrality(node) = sum(1 / distance from node to every other node excluding itself)

As with closeness centrality, we can also calculate a normalized harmonic centrality with the following formula:

normalized harmonic centrality(node) = sum(1 / distance from node to every other node excluding itself) / (number of nodes - 1)

```python
harmonic_query = """
    CALL gds.alpha.closeness.harmonic.stream({
      nodeProjection: 'User',
      relationshipProjection: 'FOLLOWS'
    })
    YIELD nodeId, centrality
    RETURN gds.util.asNode(nodeId).id AS user, centrality
    ORDER BY centrality DESC
"""
```

```python
with driver.session() as session:
    rows = session.run(harmonic_query)
    harmonic_df = pd.DataFrame([dict(record) for record in rows])
```

```python
harmonic_df
```

## Betweeness Centrality
Betweenness centrality is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. We usually use it to find bottlenecks, control points, and vulnerabilities.

The algorithm calculates unweighted shortest paths between all pairs of nodes in a graph. Each node receives a score, based on the number of shortest paths that pass through the node. Nodes that more frequently lie on shortest paths between other nodes will have higher betweenness centrality scores.

```python
between_query = """
    CALL gds.betweenness.stream('social')
    YIELD nodeId, score
    RETURN gds.util.asNode(nodeId).id AS user, score
    ORDER BY score DESC
"""
```

```python
with driver.session() as session:
    rows = session.run(between_query)
    between_df = pd.DataFrame([dict(record) for record in rows])
```

```python
between_df
```

 Alice is the main broker in this network, but Mark and Doug aren’t far behind. In the smaller subgraph all shortest paths go through David, so he is important for information flow among those nodes.

## Page Rank

PageRank is the best known of the centrality algorithms. It measures the transitive (or directional) influence of nodes. All the other centrality algorithms we discuss measure the direct influence of a node, whereas PageRank considers the influence of a node’s neighbors, and their neighbors.

```python
page_rank_query = """
    CALL gds.pageRank.stream('social')
    YIELD nodeId, score
    RETURN gds.util.asNode(nodeId).id AS user, score
    ORDER BY score DESC
"""
```

```python
with driver.session() as session:
    rows = session.run(page_rank_query)
    page_rank_df = pd.DataFrame([dict(record) for record in rows])
```

```python
page_rank_df
```

From the result above, it shows that Alice is the most influential user. 
Page Rank implementations vary, so they can produce different scoring even when the ordering is the same.
